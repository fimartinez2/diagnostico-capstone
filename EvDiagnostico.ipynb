#### Importar librerías

```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
import json
warnings.filterwarnings("ignore")
```

##### Definir Dataset

```python
with open('dataset.json') as file:
     data = [json.loads(line) for line in file.readlines()]
```

```python
df = pd.json_normalize(data)
df.head()
```

##### Funciones

```python
def most_retweeted(df,n=10):
    # Recibe dataframe y retorna los 10 tweets mas retweeteados.
    top10 = df.nlargest(n, columns=['retweetCount'])
    return top10[['content', 'retweetCount']]
    

testdf = df.sample(n=1000, random_state=1)
most_retweeted(testdf)
```

```python
def most_tweets(df,n=10):
    # Recibe dataframe y retorna los 10 usuarios con más tweets.
    return df['user.username'].value_counts()[:n]

most_tweets(testdf)
```

```python
def dias_con_mas_tweets(df,n=10):
    # Recibe dataframe y retorna dias con más tweets.
    df['dia']=df['date'].str[:9]
    return df['dia'].value_counts()[:n]
    
dias_con_mas_tweets(testdf)
```

```python
def hashtags_mas_usados(df, n=10):
    # Recibe dataframe y retorna hashtags más usados.
    return df.content.str.extractall(r'(\#\w+)')[0].value_counts()[:n]

hashtags_mas_usados(testdf)
```

##### Main

```python
def main(df, n=10):
    # recopila las funciones.
    print(f'1. Top {n} tweets mas retweeted:\n')
    print(most_retweeted(df,n))
    print(f'\n2. Top {n} usuarios con más tweets:\n')
    print(most_tweets(df,n))
    print(f'\n3. Top {n} dias con más tweets:\n')
    print(dias_con_mas_tweets(df,n))
    print(f'\n4. Top {n} hashtags más utilizados:\n')
    print(hashtags_mas_usados(df,n))
```

### Resultados

```python
main(testdf)
```
