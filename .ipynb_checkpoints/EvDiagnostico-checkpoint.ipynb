#### Importar librer√≠as

```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
import json
warnings.filterwarnings("ignore")
```

##### Definir Dataset

```python
with open('dataset.json') as file:
     data = [json.loads(line) for line in file.readlines()]
```

```python
df = pd.json_normalize(data)
df.head()
```

##### Funciones

```python
def most_retweeted(df):
    # Recibe dataframe y retorna los 10 tweets mas retweeteados.
    top10 = df.nlargest(n=10, columns=['retweetCount'])
    return top10[['content', 'retweetCount', 'user.username']]
    

testdf = df.sample(n=1000, random_state=1)
most_retweeted(testdf)
```
